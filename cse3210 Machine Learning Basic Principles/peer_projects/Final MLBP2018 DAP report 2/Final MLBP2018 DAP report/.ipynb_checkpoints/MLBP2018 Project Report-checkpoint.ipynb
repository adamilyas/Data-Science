## 3. Methods and experiments

Down below by running the code a block diagram depicting our approach can be viewed. The block diagram is only to more clearly depict our work process, but the process is also explained below. 

First, we tried our own multiclass classification with logistic regression and gradient descent. We divided the data 50/50 into training and validation data. Multiclass classification with logistic regression and gradient descent was easy to build since we had used it in this course, and only needed to change some things to get it to work with the data provided. We thought that it was a good choice for this problem. The training accuracy was about 50% which was not great, especially as the algorithm initially guessed all labels as "1", which provided approximately the same accuracy. After the first experiments, we tried to change the parameters and manually tested the algorithm. The training accuracy increased to approximately 54%.

Since the accuracy was very low with our own logistic regression and gradient descent, we decided to search for other algorithms and libraries. We found the sklearn library, where logistic regression was an inbuilt function. With this function, we reached a much better accuracy of 60%. 

In order to further improve on our results we decided to use Principal Component Analysis (PCA) to process our data. We used PCA from the Sklearn library to process the data, so that we would not need to code the math ourselves. Using PCA our results improved a little bit compared to the results without PCA. After that, we tried to use Linear Discriminant Analysis (LDA) to improve on the results, interestingly in our own tests the results did not improve, but when we returned it to the Kaggle competition the results did improve.

Finally, we wanted to test neural networks and how that would affect our results. The easiest way was to use the Sklearn library once again. After some googling we found that a Multilayer Perceptron (MLP) may be suitable for the task at hand. By using MLP and preprocessing the data using LDA, we finally reached our best result in logloss and accuracy for the training and validation data, and the best results when returned to the Kaggle competition.

```python
# Trials with ML algorithms
#Multiclass classification with logistic regression and gradient descent
def sigmoid_func(z):
    sigmoid = 1/(1+np.exp(-z))
    return sigmoid


def gradient(X,y,w):
    
    sigmoid = sigmoid_func(np.dot(X, w.T))
    N = np.size(X, 0)
  
    grad = 1/N*np.dot(np.transpose(sigmoid-y), X)
    
    return grad
    
    
def logisticRegression_func(X,y,step_size, K):
    N = X.shape[0]
    d = X.shape[1]
    # Initialize w as 1xd array.
    w = np.zeros((1,d))
    loss = float('inf')
    loss_list = []
    
    for i in range(K):
        
        grad = gradient(X, y, w)
        
        w = w-step_size*grad
        
        loss_list = np.append(loss_list, grad)
        
    return loss_list, w
    
    
def predict_label(X,w):
    
    y = sigmoid_func(np.dot(X, w.T))
    
    for i in range(0, len(X)):
        if y[i] < 0.5:
            y[i] = 0
        else:
            y[i] = 1
    
    return y


def set_labels(y, genre):
    y01 = np.zeros((len(y),1))
    
    for i in range(len(y)):
        if y[i] == genre:
            y01[i] = 1
        else:
            y01[i] = 0
    
    return y01



def do_PCA(Xtrain, Xtest, var):
    
    pca = PCA(var).fit(Xtrain)
    Xtrain = pca.transform(Xtrain)
    Xtest = pca.transform(Xtest)
    
    return Xtrain, Xtest


#
#def do_linear_regression(X, y, Xval, yval, step_size, num_iter):
#    
#    y01 = set_labels(y, 1)
#    e_list, w_opt = logisticRegression_func(X, y01 ,step_size,num_iter)
#    
#    y_predict_train = sigmoid_func(np.dot(X, w_opt.T))
#    y_predict_val = sigmoid_func(np.dot(Xval, w_opt.T))
#    
#    
#    for i in range(2, 11):
#        y01 = set_labels(y, i)
#        
#        e_list, w_opt = logisticRegression_func(X, y01 ,step_size,num_iter)
#        
#        y_predict_train = np.hstack((y_predict_train, sigmoid_func(np.dot(X, w_opt.T))))
#        
#        y_predict_val = np.hstack((y_predict_val, sigmoid_func(np.dot(Xval, w_opt.T))))
        
    
#    y_hat_train = y_predict_train.argmax(axis=1)+1
#    y_hat_val = y_predict_val.argmax(axis=1)+1
    
    
#    accuracy = calculate_accuracy(y, y_hat_train)
    
#    print ("Training data accuracy is: %f with stepsize %f" % (accuracy, step_size))
    
#    accuracy = calculate_accuracy(yval, y_hat_val)
#    print ("Validation data accuracy is: %f with stepsize %f" % (accuracy, step_size))
    
#    return
    
#num_iter = 1000
#step_sizes = [0.007]
#explained_variance = [70]

#use own logistic regression that sucks
#for step_size in (step_sizes):
#y_hat_val, y_logloss = do_linear_regression(Xtrain, ytrain, Xtest, ytest, step_size, num_iter)

    
```

```python
#Final method of genre classification. Linear Discriminant Analysis to process data and Multilayer Perceptron to classify labels.
def do_LDA(Xtrain, Xtest, ytrain):
    
    lda = LinearDiscriminantAnalysis().fit(Xtrain, ytrain)
    Xtrain = lda.transform(Xtrain)
    Xtest = lda.transform(Xtest)
    
    return Xtrain, Xtest
    
    
def MLP_classify(Xtrain, ytrain, Xtest):
    fitted = MLPClassifier(max_iter = 5000, activation = 'logistic', alpha = 0.001).fit(Xtrain, ytrain)
    y_val_pred = fitted.predict(Xtest) 
    y_logloss = fitted.predict_proba(Xtest)
    y_train_pred = fitted.predict(Xtrain) 
    y_train_logloss = fitted.predict_proba(Xtrain)
    
    
    return y_val_pred, y_logloss, y_train_pred, y_train_logloss


def calculate_accuracy(y,y_hat):

    sum = 0
    for i in range(0, len(y)):
        if y[i] == y_hat[i]:
            sum += 1
    
    accuracy = 1/len(y)*sum*100
    
    return accuracy
```

```python
#Confusion matrix 
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')



    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)
    plt.gca().invert_yaxis()
    
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()


    
```

```python
Xtrain, Xtest, ytrain, ytest, X = initialize_trainval_data()
Xtrain, Xtest = do_LDA(Xtrain, Xtest, ytrain)
y_hat_val, y_logloss, y_hat_train,y_logloss_train = MLP_classify(Xtrain, ytrain, Xtest)


print("Training logloss: %f" % log_loss(ytrain, y_logloss_train))
print("Training accuracy is %f" % calculate_accuracy(y_hat_train, ytrain))
print("Validation accuracy is %f" % calculate_accuracy(y_hat_val, ytest))
print("Best result achieved in accuracy in Kaggle = 0.65053")
print("Best result achieved in logloss in Kaggle = 0.17910")


cnf_matrix = confusion_matrix(ytest, y_hat_val)

plt.figure(figsize = (8,8))
plot_confusion_matrix(cnf_matrix, np.arange(1, 11),
                  title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure(figsize = (8,8))
plot_confusion_matrix(cnf_matrix, np.arange(1, 11), normalize=True,
                      title='Normalized confusion matrix')
```

## 5. Discussion/Conclusions

The final method of predicting the labels was using LDA (Linear Discriminant Analysis) to preprocess the data and MLP (Multilayer Perceptron) to predict the labels. This achieved the best results both in testing on the training data and when returned to the Kaggle competition.

Using accuracy to determine how good a classifier is for an imbalanced multiclass dataset may in this case not be optimal. In this case simply assigning 1 as the predicted label to every song provides quite a good accuracy, a result of a high number of songs being in the 1st genre compared to the others. A possibly better performance measure would be using the F-score. The F-score takes into account both how many of the predictions were right of all particular predictions and how many predictions of all in a particular genre were right (Sasaki 2007). This is different from the accuracy measure used which only compares how many predictions were simply right compared to all predictions.

From the results it can be seen that the training accuracy is very high and the logloss is low. In other words, the estimator is very good at identifying the genres of the different songs in the training data. However, when the estimator is used on the validation data the accuracy drops significantly. This means that the estimator may be overfitting to the test data. This huge difference in training and validation accuracy was not achieved when using logistic regression, but MLP was chosen as the final algorithm because it still provided the best validation accuracy for the same amount of data.

The confusion matrix shows that the estimator mainly thinks that songs are of genre 1. Because the majority of the songs are of genre 1, the estimator performs quite well. 

In order to improve on the scores, the data could processed more. For example, redundant features could be identified and removed. Additionally the different parameters of the classifiers could probably be optimized and examined further, as for now they were only tested manually to see if different values provide better result. 

By doing this project we have reached our goals in learning how to implement machine learning algorithms, but more work could have been done, if there had been more time. The classification could surely be made better by further examining and widening our understanding of how different algorithms really work.
