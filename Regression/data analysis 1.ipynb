# Dataset used for Linear Regression

`sns.pairplot`, `sns.jointplot`, `sns.lmplot`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
```

```python
customers = pd.read_csv("Ecommerce Customers")
customers.head()
```

```python
customers.info()
```

```python
customers.describe()
```

```python
sns.pairplot(customers)
```

```python
sns.heatmap(customers.corr(), annot=True, cmap='Blues')
```

```python
for feature in ['Length of Membership', 'Time on App', 'Avg. Session Length']:
    feature_x = customers[feature]
    feature_y = customers['Yearly Amount Spent']
    sns.jointplot(x=feature_x, y=feature_y, kind='hex')
```

```python
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=customers)
```
